# Herencia en Programación Orientada a Objetos (OOP)

La **herencia** es una relación de **especialización** y **generalización** donde una clase (subclase) adquiere atributos y métodos de otra (superclase).

## 1. Conceptos Fundamentales
- **Subclase (Hija):** La clase que hereda. Es una especialización de la superclase.
- **Superclase (Madre):** La clase de la cual se hereda. Es la generalización.
- **Reutilización:** Permite aprovechar el código ya escrito en clases superiores.
- **Relación "Es un":** Un `FurgonEscolar` **es un** `Auto`.

````python
class Auto:
    def __init__(self, marca, modelo, año):
        self.marca = marca
        self.modelo = modelo
        self.año = año
        self._kilometraje = 0

    def conducir(self, kms):
        print(f"Conduciendo {kms} kilómetros")
        self._kilometraje += kms

class FurgonEscolar(Auto):
    def __init__(self, marca, modelo, año):
        # Llamada al init de la superclase
        Auto.__init__(self, marca, modelo, año)
        self.estudiantes = []

    def inscribir_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)
````

## 2. Sobreescritura de Métodos (*Overriding*)
Ocurre cuando definimos en la subclase un método que ya existe en la superclase con el mismo nombre. Esto permite que la subclase implemente una versión especializada del comportamiento.

## 3. Uso de `super()`
Es la forma recomendada de acceder a métodos de la clase superior sin nombrarla explícitamente. Ayuda a que el código sea más mantenible y evita problemas en herencia múltiple.

````python
class FurgonEscolarMejorado(Auto):
    def __init__(self, marca, modelo, año):
        # Uso de super() en lugar del nombre de la clase
        super().__init__(marca, modelo, año)
        self.estudiantes = []

    # Overriding de conducir
    def conducir(self, distancia):
        print(f"Conduciendo con cuidado {distancia} kms, llevo niños.")
````

## 4. Herencia con Built-ins
Podemos extender clases integradas de Python como `list`, `dict` o `set` para agregarles funcionalidad personalizada.

````python
class ListaDeContactos(list):
    def buscar(self, nombre):
        # self se comporta como una lista
        return [c for c in self if nombre in c.nombre]
````

# Polimorfismo y Duck Typing en Python
---
## 1. Polimorfismo
Es la propiedad de enviar mensajes sintácticamente iguales a objetos de tipos distintos, utilizando una misma **interfaz**.

### A. Overriding (Sobreescritura)
Ocurre cuando una subclase implementa un método que ya existe en su superclase para darle un comportamiento específico.

**Ejemplo:**
Una clase `Variable` define `representante()`. Las subclases `Ingresos` (promedio) y `Comunas` (moda) sobreescriben este método.

````python
import statistics

class Variable:
    def __init__(self, data):
        self.data = data
    def representante(self):
        pass

class Ingresos(Variable):
    def representante(self):
        return statistics.mean(self.data)

# Se invoca el mismo método en distintos tipos de objeto
p = Ingresos()
print(f"Representante Ingresos: {p.representante()}")
````

### B. Overloading (Sobrecarga)
* **En Funciones:** Capacidad de definir un método con el mismo nombre pero distintos parámetros. **Python NO soporta overloading de funciones** directamente; se queda siempre con la definición más reciente.
* **En Operadores:** Python permite personalizar el comportamiento de operadores *built-in* (como `+`, `<`, `==`) mediante métodos especiales.

**Métodos comunes de Overloading de Operadores:**
* `__add__`: Operador `+`
* `__lt__`: Operador `<` (less than)
* `__eq__`: Operador `==` (equal)
* `__le__`, `__gt__`, `__ge__`, `__ne__`: Otros operadores de comparación.

## 2. Representación de Objetos: `__str__` vs `__repr__`
Ambos devuelven un *string*, pero con objetivos diferentes:

| Método | Objetivo | Dirigido a... |
| :--- | :--- | :--- |
| `__str__` | Legible (*human-readable*) | Usuario final |
| `__repr__` | Completa y sin ambigüedades | Desarrollador |

*Dato clave:* Si no existe `__str__`, `print()` usará `__repr__`. Además, `__repr__` se usa para mostrar objetos dentro de una lista.

## 3. Duck Typing
> "If it walks like a duck and quacks like a duck then it is a duck".

Es una característica de lenguajes con **tipado dinámico** como Python. No importa la herencia o el tipo de clase; si el objeto tiene los métodos necesarios, Python los ejecutará.

**Ejemplo:** Una función `activar(obj)` que llama a `obj.gritar()`. Funcionará tanto si `obj` es un `Pato` como si es una `Persona`, siempre que ambos tengan el método `gritar()`.

# Multiherencia en Python

## 1. Conceptos Básicos

- **Herencia Simple:** Una subclase hereda de una única superclase (ej. `Perro` hereda de `Animal`). La especialización ocurre en una cadena vertical.
- **Multiherencia:** Una subclase hereda atributos y comportamientos de **más de una superclase** simultáneamente. Permite mezclar funcionalidades de distintas jerarquías.

### Ejemplo de estructura:
```python
class Docente: 
    pass
class Investigador: 
    pass
class Academico(Docente, Investigador): 
    pass
```

## 2. Ventajas, Desafíos y Uso Adecuado

| Aspecto | Descripción |
| :--- | :--- |
| **Ventajas** | Flexibilidad, reutilización de código y representación natural de múltiples roles (ej. un Académico es Docente e Investigador). |
| **Desafíos** | Ambigüedad en nombres de métodos, complejidad en el diseño y necesidad de entender el Orden de Resolución de Métodos (MRO). |
| **Uso Adecuado** | Usar solo cuando aporte claridad. No forzarla solo para reutilizar código; considerar **composición** como alternativa. |

## 3. Resolución de Conflictos y MRO

Cuando dos superclases tienen un método con el mismo nombre, Python utiliza el **MRO (Method Resolution Order)** basado en el algoritmo **C3 Linearization**.

**Orden de búsqueda:**
1. La propia clase derivada.
2. La **primera** clase base declarada en la definición.
3. Las siguientes clases base en el orden de declaración.

Si definimos `class Academico(Docente, Investigador)`, Python buscará primero en `Docente`. Si invertimos el orden a `(Investigador, Docente)`, buscará primero en `Investigador`.

## 4. El Problema del Diamante

Ocurre cuando hay más de un camino en la jerarquía desde una subclase hacia una superclase común. 

**Riesgo:** Si no se maneja correctamente (por ejemplo, llamando manualmente a las clases base como `Vehiculo.acelerar(self)`), los métodos de la clase superior pueden ejecutarse **múltiples veces**, causando efectos secundarios no deseados como contadores duplicados o inicializaciones dobles.

# Clases Abstractas en Python



## 1. Conceptos Fundamentales

### ¿Qué es una Clase Abstracta?
Desde el punto de vista del modelamiento, una clase es **abstracta** si:
* **No está diseñada para ser instanciada:** No se deben crear objetos directamente de ella.
* **Sirve de base:** Su propósito es ser heredada por otras clases concretas.
* **Define un contrato:** Establece comportamientos mínimos (métodos) que sus subclases **deben** implementar.

**Ejemplo de modelamiento:** La clase `Mamifero` es abstracta. No existe un animal que sea "solo un mamífero"; debe ser algo concreto como un `Perro` o una `Ballena`.

### Métodos Abstractos y Normales
* **Métodos Abstractos:** Representan comportamientos obligatorios que difieren en su implementación según la subclase.
* **Métodos Normales:** Pueden contener lógica ya implementada que las subclases heredan directamente para evitar repetición de código.

## 2. Implementación en Python: Módulo `abc` 

Python requiere el módulo `abc` (*Abstract Base Class*) para definir estas estructuras.

* **`ABC`**: Clase de la cual se debe heredar para marcar la clase como abstracta.
* **`@abstractmethod`**: Decorador para definir métodos que las subclases están obligadas a sobreescribir.

````python
from abc import ABC, abstractmethod

class Base(ABC):
    def __init__(self, nombre):
        self.nombre = nombre

    @abstractmethod
    def metodo_obligatorio(self):
        # Las subclases deben implementar esto
        pass

    def metodo_comun(self):
        # Lógica compartida por todas las subclases
        print(f"Hola, soy {self.nombre}")
````

## 3. Reglas de Instanciación

1.  **Error de Instanciación:** Si intentas hacer `instancia = Base()`, Python lanzará un error porque contiene métodos abstractos sin implementar.
2.  **Implementación Parcial:** Si una subclase hereda de una clase abstracta pero **no implementa todos** los métodos abstractos, esa subclase también será considerada abstracta y no podrá instanciarse.
3.  **Uso de `super()`:** Los métodos abstractos en la clase base **pueden tener lógica**. Una subclase puede implementar el método y llamar a la versión de la base usando `super().metodo()`.

# Decoradores y Propiedades

## 1. Conceptos Fundamentales

### Objetos de Primera Clase
En Python, las funciones son **objetos de primera clase**, lo que permite:
* Asignarlas a variables.
* Pasarlas como argumentos a otras funciones.
* Retornarlas desde otras funciones.
* Almacenarlas en estructuras de datos.

**Nota crucial:** 
* `mi_funcion`: Referencia a la función (no se ejecuta).
* `mi_funcion()`: Ejecución de la función y referencia a su valor de retorno.

### Decoradores
Un **decorador** es una función que recibe otra función como entrada y retorna una versión modificada de la misma. Permite añadir funcionalidades sin cambiar el código original.

**Sintaxis:**
```python
@decorador
def mi_funcion():
    pass
```
Es equivalente a: `mi_funcion = decorador(mi_funcion)`

## 2. Decoradores Integrados (Built-in)

### @property (Getters y Setters)
Permite manejar atributos protegidos (por convención con `_`). Se utiliza para:
* **Getter:** Acceder a un método como si fuera un atributo (`@property`).
* **Setter:** Validar o modificar datos antes de asignarlos (`@nombre.setter`).

### @staticmethod
Define métodos que pertenecen a la clase pero no requieren acceso a la instancia (`self`). Son útiles para organizar funciones relacionadas con la clase que actúan de forma independiente al estado del objeto.

````python
class CuentaBancaria:
    def __init__(self, titular, saldo_inicial):
        self.titular = titular
        self._saldo = saldo_inicial  # Atributo protegido

    @property
    def saldo(self):
        """Getter: Devuelve el saldo actual."""
        return f"${self._saldo:.2f}"

    @saldo.setter
    def saldo(self, valor):
        """Setter: Valida que el depósito no sea negativo."""
        if valor < 0:
            print("Error: El saldo no puede ser negativo.")
        else:
            self._saldo = valor

# Uso
mi_cuenta = CuentaBancaria("Jose", 1000)

print(mi_cuenta.saldo)  # Acceso como atributo -> $1000.00

mi_cuenta.saldo = 1500  # Actualización exitosa
mi_cuenta.saldo = -50   # Muestra error: El saldo no puede ser negativo.
````

````python
import math

class Circulo:
    def __init__(self, radio):
        self._radio = radio

    @property
    def radio(self):
        return self._radio

    @radio.setter
    def radio(self, valor):
        if valor <= 0:
            raise ValueError("El radio debe ser mayor a cero.")
        self._radio = valor

    @property
    def area(self):
        """Getter para un atributo calculado (solo lectura)."""
        return math.pi * (self._radio ** 2)

# Uso
c = Circulo(5)
print(f"Radio: {c.radio}") # 5
print(f"Área: {c.area:}") # 78.54

c.radio = 10  # Cambiamos el radio usando el setter
print(f"Nueva área: {c.area:}") # 314.16
````

1. Una `@property` para leer la temperatura.

3. Un método estático `@staticmethod` que convierta Celsius a Fahrenheit ($F = C * 1.8 + 32$).

````python
class Termometro:
    def __init__(self, temperatura):
        self._celsius = temperatura

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, valor):
        if valor < -273.15:
            print("¡Error! Por debajo del cero absoluto.")
        else:
            self._celsius = valor

    @staticmethod
    def a_fahrenheit(c):
        return c * 1.8 + 32

# Pruebas
t = Termometro(25)
print(f"Celsius: {t.celsius}")
print(f"En Fahrenheit: {Termometro.a_fahrenheit(t.celsius)}")
t.celsius = -300 # Debería mostrar error
````

# Diagramas de Clases (UML)

El **diagrama de clases** es una herramienta del *Lenguaje de Modelado Unificado* (UML) que permite visualizar la estructura de un sistema, detallando sus clases, atributos, métodos e interacciones.

---

## 1. Elementos de un Diagrama

### Clases
Se representan como rectángulos divididos en tres niveles:
1.  **Nombre de la clase.**
2.  **Atributos:** Variables propias. En este curso, se incluyen las *properties* por convención.
3.  **Métodos:** Funciones o comportamientos. Se debe especificar nombre, parámetros y tipo de retorno.

### Relaciones
Definen cómo interactúan los objetos:

* **Contención (⧫/◊):** Un objeto tiene a otro como atributo.
    * *Composición (⧫):* El objeto contenido depende de la existencia del contenedor (ej: Mario en el Juego).
    * *Agregación (◊):* El objeto contenido es independiente (ej: Goombas que pueden existir fuera de un líder específico).
    * **Nota:** En este curso no se hace distinción obligatoria entre ambas al usar el rombo.

* **Herencia (ᐅ):** Relación de especialización/generalización. La flecha apunta hacia la **superclase**.
    * Las subclases heredan atributos y métodos implícitamente, por lo que no se repiten en el diagrama.
