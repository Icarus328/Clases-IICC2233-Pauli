# Paradigma de Programación Funcional


## 1. Paradigmas de Programación

Un paradigma es un enfoque o estrategia para enfrentar problemas. Python es **multiparadigma**, permitiendo mezclar estos estilos:

- **Procedimental**: Instrucciones lineales paso a paso.
- **Vectorial**: Paralelismo implícito en cálculos matemáticos.
- **Declarativa**: Se describe el "qué" (problema) y el computador decide el "cómo" (ej. SQL).
- **Orientada a Objetos (OOP)**: Modelado mediante objetos y sus interacciones.
- **Funcional**: Programación de alto nivel donde la solución es un conjunto de funciones sin estado.

## 2. Programación Funcional y Funciones Puras

En este paradigma, las funciones se comportan como en matemáticas: para un mismo valor de entrada, siempre se obtiene el mismo valor de salida.

### ¿Qué es una Función Pura?
Una función es **pura** si cumple con dos condiciones:
1. **Determinismo**: El retorno depende exclusivamente de los parámetros de entrada.
2. **Sin efectos secundarios (side effects)**: No modifica variables externas, no realiza `print`, no escribe archivos ni altera el estado global.

### Ejemplos de Pureza

```python
from copy import deepcopy
from random import Random

# EJEMPLO IMPURO: Modifica una lista externa (side effect)
lista_global =
def agregar_impuro(n):
    lista_global.append(n)

# EJEMPLO PURO: No toca el exterior, retorna una copia nueva
def agregar_puro(lista, n):
    nueva_lista = deepcopy(lista)
    nueva_lista.append(n)
    return nueva_lista

# EJEMPLO PURO CON RANDOM: Usa una semilla local para ser determinista
def aleatorio_puro(seed_val):
    generador = Random(seed_val)
    return generador.random()
```

## 3. Consecuencias del Paradigma Funcional

1. **Claridad y Depuración**: Si un valor cambia inesperadamente, sabemos que no fue una función pura.
2. **Memoización (Cache)**: Si el programa recuerda el input, no necesita re-calcular la función (ej. usando `@cache`).
3. **Flujos de Datos (Data Flow)**: Se piensa en transformar datos en lugar de almacenar variables.
4. **Modularidad**: Las iteraciones se vuelven más amigables de modelar.

```python
from functools import cache

@cache
def fibonacci_eficiente(n):
    if n < 2:
        return n
    return fibonacci_eficiente(n-1) + fibonacci_eficiente(n-2)

print(f"Fibonacci de 50: {fibonacci_eficiente(50)}")
```

# Funciones Generadoras en Python

## 1. Conceptos Fundamentales

* **Generadores:** Permiten iterar sobre datos sin almacenarlos en estructuras especiales, evitando el uso innecesario de memoria.
* **Sentencia `yield`:** Es análoga a `return`, pero con una diferencia clave: cede el control temporalmente y asegura que, en la próxima llamada, la ejecución parta desde donde se dejó.
* **Retorno de la función:** Al llamar a una función generadora, esta no se ejecuta de inmediato, sino que retorna un **objeto generador**.

### Ejemplo Básico: Conteo Decreciente


```python
from typing import Generator

def conteo_decreciente(n: int) -> Generator:
    while n > 0:
        yield n
        n -= 1

x = conteo_decreciente(3)
print(next(x))  # 3
print(next(x))  # 2
```

## 2. Casos de Uso Comunes

* **Series infinitas:** Como la serie de Fibonacci, que puede generar elementos indefinidamente sin agotar la memoria.
* **Lectura de archivos grandes:** Utilizar un generador para leer línea por línea evita cargar archivos pesados (como un `.csv`) completamente en la RAM.
* **Operaciones acumulativas:** Por ejemplo, calcular el máximo visto hasta el momento en una lista.

## 3. Comunicación Bidireccional (`send`)

El método `.send(valor)` permite enviar datos de vuelta a la función generadora.

* La expresión `valor_recibido = yield contador` captura lo que se envía desde fuera.
* **Regla de oro:** No se puede enviar un valor no nulo a un generador que recién comienza; primero se debe avanzar hasta el primer `yield` usando `next()` o `.send(None)`.

## 4. Delegación de Generadores (`yield from`)

La sentencia `yield from` permite crear un generador a partir de otro iterable (como una lista o un `range`) o incluso de otro generador. Esto resulta en un código más limpio y modular.

## 5. Definición Técnica

Para que Python considere que una función es generadora, basta con que **contenga la palabra clave `yield`** en cualquier parte de su cuerpo, incluso si el código es inalcanzable (por ejemplo, después de un `return` o dentro de un `if False`).
