# Diccionarios

Los diccionarios son una estructura de datos mutable y no secuencial, donde llamamos a un valor "key" para obtener un valor "value".

Actuan de forma similar a una función, donde la llave sería el input, y el valor el output. Se construyen así:

```python
monedas = {
    "Chile": "Peso",
    "Perú": "Soles",
    "España": "Euro", 
    "Holanda": "Euro",
    "Brasil": "Real"
}
```

Podemos obtener sus values de dos maneras.

1) llamando a diccionario[key]

```python
print(monedas["Chile"])
print(monedas["Perú"])
```

2. llamando a la funcion get, llamando a diccionario.get(a, b). Donde a es la key cuyo value buscamos, y b lo que retorna la funcion si la key "a" no existe.

```python
print(monedas.get("Chile", "no hay plata"))
print(monedas.get("Brasil", "no hay plata"))
```

Podemos añadir pares key/value a un diccionario, o cambiar el value asignado a una key de la siguiente forma.

```python
monedas["Chile"] = "escudos"
monedas["Brasil"] = "reales"
```

```python
print(monedas.get("Chile", "no hay plata"))
print(monedas.get("Brasil", "no hay plata"))
```

Tanto las keys como los values pueden ser de cualquier tipo, no es necesario que sean todos ints o todos strings

```python
monedas[3.14] = ("Peso", "Dolar")
monedas
#aqui podemos ver 1 key que es int y todas las demas strings, y un value que es una tuple.
```

Podemos eliminar pares key/value con esta sintaxis

```python
del monedas[3.14]
monedas
```

Es necesario que la key exista, y no pueden existir keys sin value, ni values sin keys.

Podemos usar la funcion "in" para saber si una key está en un diccionario. Devuelve un booleano.

```python
print('Chile' in monedas)
print("Alemania" in monedas)
```

Un value en un diccionario puede ser cualquier cosa, pero la key solo puede ser un elemento inmutable (como las tuplas, los strings o los ints) y en la practica, solo serán strings o ints. También, se pueden repetir values, pero jamás se pueden repetir keys.

Podemos usar la funcion keys con diccionario.keys() para obtener todas las keys

```python
print(monedas.keys())
```

Podemos usar la funcion values con diccionario.values() para obtener todos los values

```python
print(monedas.values())
```

Podemos usar la funcion items con diccionario.items() para obtener todos los pares key/value

```python
print(monedas.items())
```

Finalmente, es posible usar funciones para crear diccionarios

```python
paises = ['Chile', 'Perú', 'España', 'Holanda', 'Brasil']
monedas = ['escudos', 'Soles', 'Euro', 'Euro', 'reales']

diccionario_monetario = {}

for i in range(len(paises)):
    diccionario_monetario[paises[i]] = monedas[i]

print(diccionario_monetario)
```

Es necesario crear el diccionario vacio primero.

# Sets

Los sets son una estructura de datos, similar a una lista, que no admite duplicados, sirven para revisar si un valor se encuentra dentro de ellos.

Es posible crearlos a partir de listas, eliminando los duplicados, o crearlos desde 0

```python
lista_artistas = ["Olivia Newton-John", "Daddy Yankee", "Sting", 
                  "Dream Theater", "Mon Laferte", "Sting"]
print(set(lista_artistas))

conjunto_artistas = {"Olivia Newton-John", "Daddy Yankee", "Sting", 
                     "Dream Theater", "Mon Laferte", "Sting"}
print(conjunto_artistas)
```

Los elementos dentro de un set pueden ser de cualquier tipo (ints, strings, instancias, etc).

Como puedes ver, el orden del print es distinto en ambos casos, a pesar de que el input desde el que son creados son el mismo, esto es por que los sets no tienen ningún tipo de orden, y no se pueden acceder por indice.

```python
print(conjunto_artistas[2])
```

Los sets soportan distintas operaciones

```python
#Podemos añadir elementos a los sets

conjunto_artistas.add("Taylor Swift")
print(conjunto_artistas)
```

```python
#Podemos quitar elementos de los sets

conjunto_artistas.remove("Daddy Yankee")
print(conjunto_artistas)
```

```python
#Esta es otra forma de quitar elementos de los sets, pero no da un error si el elemento no existe

conjunto_artistas.discard("The Beatles")
print(conjunto_artistas)
```

```python
#Se puede iterar con for 

for artista in conjunto_artistas:
    print(f"Por favor, ¡saluden a {artista}!")
```

```python
#podemos verificar si hay elementos dentro de un set

print("Sting" in conjunto_artistas)
```

Los sets pueden ser vistos como conjuntos matematicos, y soportan operaciones de conjuntos

```python
set_a = {0, 1, 2, 3}
set_b = {5, 4, 3, 2}

#Union

set_union = set_a.union(set_b)
print(set_union)
```

```python
#Interseccion

set_intersection = set_a.intersection(set_b)
print(set_intersection)
```

# Args y Kwargs

Al crear una funcion, decides los inputs que recibe y el orden en que lo hace

```python
def suma(a, b):
    return a + b

print(suma(2, 5))
```

Pero con la sintaxis correcta, tambien puede recibir listas, tuplas o diccionarios

```python
lista = [2, 5]
tupla = (2, 5)
diccionario = {"a":2, "b":5}

print(suma(*lista))
print(suma(*tupla))
print(suma(**diccionario))
```

Recuerdas que en las clases, podiamos poner valores "por defecto" a algunos atributos? también se puede hacer con funciones!

```python
def suma(a, b=5):
    return a + b

print(suma(2))
```

Pero si especificamos el atributo, funciona de manera normal

```python
print(suma(2, 3))
```

En el caso de que hayan multiples atributos a hacer funcionar, debemos especificar los que queremos modificar.

```python
def suma(a, b = 5, c = 7):
    return a + b + c
```

```python
print(suma(2))
```

```python
print(suma(3, 2)) #c ) 7
```

```python
print(suma(3, c = 8)) #b = 5
```

```python
print(suma(3, c = 3, b = 2))
```

```python
print(suma(*[3, 2]))
```
