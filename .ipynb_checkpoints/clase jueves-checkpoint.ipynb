## 1. Herencia

La **herencia** (*inheritance*) corresponde a una relación de **especialización** y **generalización** entre clases. Una **subclase** hereda atributos y métodos de una **superclase**.

### Ejemplo: Clase Auto y Furgón Escolar

```python
class Auto:
    def __init__(self, marca, modelo, año, color, km):
        self.marca = marca
        self.modelo = modelo
        self.año = año
        self.color = color
        self._kilometraje = km
        self._dueño = None

    def conducir(self, kms):
        print(f"Conduciendo {kms} kilómetros")
        self._kilometraje += kms

    def leer_odometro(self):
        return self._kilometraje

class FurgonEscolar(Auto):
    def __init__(self, marca, modelo, año, color, km):
        # Llamada explícita al constructor de la superclase
        Auto.__init__(self, marca, modelo, año, color, km)
        self.estudiantes = []

    def inscribir_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)
```

### Sobreescritura de métodos (*Overriding*) y `super()`
El *overriding* permite que una subclase implemente una versión especializada de un método heredado. El método `super()` permite invocar métodos de la superclase sin nombrarla explícitamente.

```python
class FurgonEscolarEspecializado(Auto):
    def __init__(self, marca, modelo, año, color, kms):
        super().__init__(marca, modelo, año, color, kms)
        self.estudiantes = []

    def conducir(self, distancia):
        # Sobreescritura del método conducir
        print(f"Conduciendo con cuidado {distancia} kilómetros")
```

## 2. Polimorfismo

El **polimorfismo** es la propiedad de enviar mensajes sintácticamente iguales a objetos de tipos distintos.

- **Overriding**: Redefinir un método en la subclase.
- **Overloading**: Definir un método con el mismo nombre pero distintos parámetros. **Python no soporta overloading de funciones**, pero sí de operadores *built-in* como `__add__` o `__lt__`.

```python
class Carro:
    def __init__(self, pan, leche):
        self.pan = pan
        self.leche = leche
    
    def __add__(self, otro): # Overloading del operador +
        return Carro(self.pan + otro.pan, self.leche + otro.leche)

    def __str__(self):
        return f"Pan: {self.pan}, Leche: {self.leche}"
```

### Duck Typing
"Si camina como pato y suena como pato, entonces es un pato". En Python, el tipo de un objeto importa menos que los métodos que define.

```python
class Pato:
    def gritar(self): print("Quack!")
class Persona:
    def gritar(self): print("¡Ahhh!")

def activar(entidad):
    entidad.gritar() # No importa si es Pato o Persona
```

## 3. Multiherencia y el Problema del Diamante

La **multiherencia** permite que una clase herede de más de una superclase. Python utiliza el **MRO (Method Resolution Order)** para resolver qué método ejecutar en caso de conflictos.

### El Problema del Diamante
Ocurre cuando dos clases heredan de una base común y luego una cuarta clase hereda de ambas, lo que puede causar llamadas duplicadas a la clase base.

```python
class Vehiculo:
    def acelerar(self):
        print("Vehículo acelerando")

class Terrestre(Vehiculo):
    def acelerar(self):
        print("Terrestre acelerando")
        super().acelerar()

class Acuatico(Vehiculo):
    def acelerar(self):
        print("Acuático acelerando")
        super().acelerar()

class Anfibio(Terrestre, Acuatico):
    def acelerar(self):
        super().acelerar()
```

## 4. Clases Abstractas

Las clases abstractas no están diseñadas para ser instanciadas, sino para servir de base a otras clases, estableciendo un "contrato" de métodos que deben ser implementados. Se utiliza el módulo `abc`.

```python
from abc import ABC, abstractmethod

class Personaje(ABC):
    @abstractmethod
    def moverse(self):
        pass

class Jugador(Personaje):
    def moverse(self):
        print("Jugador moviéndose")
```

## 5. Decoradores y Properties

Un decorador es una función que toma otra función y retorna una versión modificada de esta. Las funciones en Python son **objetos de primera clase**.

### Properties (`@property`)
Permiten manejar atributos como si fueran públicos, pero con lógica de validación interna (*getters* y *setters*).

```python
class Email:
    def __init__(self, address):
        self._email = address
    
    @property
    def email(self):
        return self._email
    
    @email.setter
    def email(self, valor):
        if "@" in valor:
            self._email = valor
        else:
            print("Email inválido")
```

## 6. Diagrama de Clases (UML)

Permite visualizar la estructura del sistema.
- **Rectángulo**: Clase (Nombre, Atributos, Métodos).
- **Flecha blanca (ᐅ)**: Herencia.
- **Rombo (⧫)**: Contención (Composición/Agregación).
