# Excepciones en Python

Una **excepción** es una condición anómala o inesperada que ocurre durante la ejecución de un programa e interrumpe su curso habitual. En Python, las excepciones son objetos que heredan de la clase `Exception`.

### Conceptos Clave
* **Runtime Exceptions:** En Python (lenguaje interpretado), las excepciones se producen en tiempo de ejecución. Si el flujo del programa no llega a la instrucción errónea, la excepción no se genera.
* **Robustez:** Un programa es robusto cuando es capaz de recuperarse ante errores inesperados o entradas inválidas mediante el **manejo de excepciones** (`raise` y `catch`).

---

## Catálogo de Excepciones Comunes

| Excepción | Descripción | Ejemplo |
| :--- | :--- | :--- |
| `SyntaxError` | Sentencia mal escrita que viola reglas sintácticas. Se detecta antes de ejecutar. | `print "Hola"` (sin paréntesis en Python 3). |
| `IndentationError`| No se respeta la indentación obligatoria de un bloque (hereda de `SyntaxError`). | Un `if` sin contenido indentado. |
| `NameError` | Intento de acceder a un nombre (variable/función) no definido. | `print(variable_inexistente)`. |
| `ZeroDivisionError`| El denominador de una división es cero. | `5 / 0`. |
| `IndexError` | Acceso a un índice fuera del rango de una secuencia. | `lista =; lista`. |
| `KeyError` | Uso de una llave (*key*) inexistente en un diccionario. | `dict = {}; dict["a"]`. |
| `AttributeError` | Uso incorrecto de métodos o atributos que no pertenecen al objeto. | Llamar `stop()` en una clase que no lo define. |
| `TypeError` | Operación con un argumento de tipo incorrecto. | `"hola" + 5` o `5()` (un entero no es llamable). |
| `ValueError` | El tipo es correcto, pero el **valor** no es apropiado para la operación. | `int("abc")` o `"".split("")`. |

> **Nota sobre `TypeError` vs `ValueError`**: `TypeError` es un caso específico de `ValueError`. Se prefiere `TypeError` cuando el problema es estrictamente el tipo de dato, ya que aporta más información para el manejo del error.

# Levantamiento y Manejo de Excepciones

## 1. Levantando excepciones: `raise`

Podemos generar una excepción manualmente creando una instancia de una clase de excepción y usando la sentencia `raise`. Esto es útil para validar datos antes de procesarlos.

### Conceptos clave:
* **Interrupción:** Las excepciones interrumpen todo el flujo del programa.
* **Propagación:** Si no se captura, la excepción sube por la pila de llamadas hasta llegar al sistema operativo.
* **Mensajes personalizados:** Se puede incluir un mensaje explicativo al instanciar la excepción.

```python
# Ejemplo de validación en un inicializador
class Circle:
    def __init__(self, center):
        if not isinstance(center, tuple):
            raise TypeError("The 'center' parameter must be a tuple")
        self.center = center

try:
    c = Circle() # Esto levantará un TypeError
except TypeError as e:
    print(f"Error capturado: {e}")
```

## 2. Manejo de Excepciones: `try` y `except`

Para evitar que el programa se detenga bruscamente, usamos bloques `try` y `except`.

### Reglas de funcionamiento:
* **Scope:** El código que podría fallar se coloca dentro del `try`.
* **Salto inmediato:** Si ocurre un error, el flujo salta al `except` y **NO regresa** a la línea que falló.
* **Errores no capturables:** `SyntaxError` o `IndentationError` no se pueden capturar en el mismo archivo porque ocurren durante la lectura, no la ejecución. (A menos que se importen desde otro archivo).

### Manejo múltiple
Podemos manejar distintos errores de forma separada o agrupada:

1.  **Bloques separados:** Para dar un tratamiento distinto a cada error.
2.  **Agrupados en tupla:** Para dar el mismo tratamiento a varios errores, e.g., `except (ZeroDivisionError, TypeError):`.

## 3. Flujos complementarios: `else` y `finally`

La estructura completa incluye dos bloques adicionales:

| Bloque | Cuándo se ejecuta |
| :--- | :--- |
| `else` | **Sólo si NO** se lanzó ninguna excepción en el `try`. |
| `finally` | **Siempre**, sin importar si hubo error o no. |

> **Nota:** `finally` es ideal para rutinas de limpieza, como cerrar archivos o conexiones a bases de datos.

## 4. Observaciones: ¿Por qué usar excepciones?

Es más recomendable usar excepciones que `if/else` para el control de errores por:
* **Obligatoriedad:** El programador no puede ignorar una excepción como sí puede ignorar un código de retorno.
* **Limpieza:** El código es más legible.
* **Mantenibilidad:** Evita crear estructuras de control personalizadas para cada función.
* **Seguridad:** Evita exponer trozos de código interno al usuario final en caso de falla.

# Iterables, Iteradores y Generadores


## 1. Iterables e Iteradores

### Definiciones
* **Iterable:** Es cualquier objeto sobre el cual se puede iterar (recorrer elementos paulatinamente). En Python, aparecen al lado derecho de un `for ... in iterable`.
* **Iterador:** Es el objeto que realiza la acción de iterar. Es retornado por el método `__iter__` del iterable y debe implementar el método `__next__`.

### Cómo hacer un objeto Iterable
Para que una clase sea iterable, debe implementar al menos uno de estos métodos:
1.  **`__iter__()`**: Retorna un objeto iterador.
2.  **`__getitem__()`**: Permite el acceso por índice. Debe levantar `IndexError` al final.

### Características del Iterador
* Usa `__next__()` para entregar el siguiente valor.
* Levanta `StopIteration` cuando no quedan más elementos.
* **Importante:** Un iterador solo se puede recorrer **una vez**, mientras que un iterable se puede recorrer múltiples veces (porque cada vez genera un iterador nuevo).

```python
# Ejemplo rápido: Iterador manual
conjunto = {10, 20, 30}
it = iter(conjunto)

print(next(it)) # 10
print(next(it)) # 20
print(next(it)) # 30
# next(it) levantaría StopIteration aquí
```

## 2. Generadores

Son un caso especial de iteradores. Permiten iterar sobre secuencias sin almacenarlas íntegramente en memoria.

### Ventajas
* **Eficiencia de memoria:** Generan elementos "on the fly" (al momento). Por ejemplo, un generador de un millón de elementos ocupa los mismos bytes que uno de diez.
* **Sintaxis:** Se crean de forma similar a las listas por comprensión pero usando paréntesis `()`.

### Lógica de filtrado vs transformación
* **Filtrar:** `(i for i in range(10) if i < 5)` -> Solo genera si cumple la condición.
* **Transformar:** `(i if i < 5 else 0 for i in range(10))` -> Modifica el valor según la condición.

```python
from sys import getsizeof

gen = (i**2 for i in range(1000000))
lista = [i**2 for i in range(1000000)]

print(f"Memoria Generador: {getsizeof(gen)} bytes")
print(f"Memoria Lista: {getsizeof(lista)} bytes")
```

# Estructuras de Datos y Listas Ligadas


## 1. El Nodo

El **nodo** es la unidad indivisible base de las estructuras de datos. Contiene:
* **Datos**: El valor almacenado.
* **Referencias**: Conexiones a otros nodos (vecinos).

En una lista ligada, cada nodo referencia a un único **sucesor** (`siguiente`).

```python
from typing import Any

class Nodo:
    def __init__(self, valor: Any = None) -> None:
        self.valor = valor
        self.siguiente = None

    def __repr__(self) -> str:
        return f"Nodo[{self.valor}]"
```

## 2. Listas Ligadas (Linked Lists)

Es una estructura secuencial donde los nodos se conectan uno tras otro. Conceptos clave:
* **Cabeza (head)**: El primer nodo de la lista.
* **Cola (tail)**: El último nodo de la lista (su `siguiente` es `None`).

### Operaciones Principales:

| Operación | Descripción | Pasos clave |
| :--- | :--- | :--- |
| **agregar(valor)** | Añade al final | Crear nodo -> conectar a la cola actual -> actualizar la cola. |
| **obtener(pos)** | Busca por índice | Recorrer secuencialmente desde la cabeza hasta la posición. |
| **insertar(val, pos)** | Inserta en cualquier punto | Buscar nodos pre y post -> reasignar referencias de `siguiente`. |

```python
class ListaLigada:
    def __init__(self) -> None:
        self.cabeza = None
        self.cola = None
        self.largo = 0

    def agregar(self, valor: Any) -> None:
        nuevo = Nodo(valor)
        if self.cabeza is None:
            self.cabeza = nuevo
            self.cola = nuevo
        else:
            self.cola.siguiente = nuevo
            self.cola = nuevo
        self.largo += 1

    def obtener(self, posicion: int) -> Any:
        if posicion < 0: return None
        nodo_actual = self.cabeza
        for _ in range(posicion):
            if nodo_actual: 
                nodo_actual = nodo_actual.siguiente
        return nodo_actual.valor if nodo_actual else None

    def __repr__(self) -> str:
        nodos = []
        act = self.cabeza
        while act:
            nodos.append(str(act.valor))
            act = act.siguiente
        return " → ".join(nodos)
```

## 3. Iteración y Protocolos

Para usar `for` en una `ListaLigada`, se deben implementar:
1.  **Iterador**: Una clase con `__next__` que mantiene el estado del recorrido.
2.  **Iterable**: La clase de la lista con el método `__iter__` que retorna el iterador.
