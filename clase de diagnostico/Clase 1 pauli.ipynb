Tipos de datos

```python
# int -> números enteros

a = 5

# float -> números decimales

b = 5.5

# string -> texto, debe estar rodeado de comillas

c = "Hola mundo"

# Boolean -> tiene dos valores posibles verdadero o falso

bool_true = True
bool_false = False

# listas -> contienen otros tipos de datos, ¡de cualquier tipo!

lista = [a, b, c, bool_true]
print(lista)

# Puedes imprimir elementos determinados de la lista, o en orden determinado
print(lista[3])
print(lista[::-1])

# Puedes añadir elementos a la lista al final
lista.append(bool_false)
print(lista)

# Puedes añadir elementos a la lista segun indice
lista.insert(1, "hola mundo, pero antes en la lista")
print(lista)

# Y lo mismo con quitar elementos
lista.remove(5) #quita la primera instancia de "5"
lista.pop(0) #quita el primer elemento
print(lista)
```

Funciones

```python
# funciones -> nos permiten tomar un bloque de codigo, y ejecutarlo sin tener que escribirlo constantemente

def sumar(numero1, numero2):
    #suma dos numeros
    suma = numero1 + numero2

    # Retornamos el resultado de la suma
    return suma

# Para llamar a la funcion, debemos asignar una variable a su "return"

resultado = sumar(2, 3)
print(resultado)
```

Archivos

```python
# Podemos leer y escribir archivos con python

# Definimos el nombre del archivo
nombre_archivo = "saludo.txt"

# Abrimos el archivo en modo escritura ('w'), sintaxis para crear archivo
with open(nombre_archivo, "w", encoding="utf-8") as archivo:
    # Escribimos el contenido
    archivo.write("hola mundo")
```

```python
nombre_archivo = "saludo.txt"

# Abrimos el archivo en modo append ('a'), sintaxis para añadir "datos" al archivo
with open(nombre_archivo, "a", encoding="utf-8") as archivo:
    # Agregamos un alto de línea y el nuevo texto
    archivo.write("\nbuen día")
# Lo corremos, revisamos, y lo corremos nuevamente
```

```python
# Cuidado con la sintaxis a usar, eliminamos el archivo :(
nombre_archivo = "saludo.txt"

with open(nombre_archivo, "w", encoding="utf-8") as archivo:
    archivo.write(":(")
```

Clases (objetos)

```python
# Una clase es una "funcion" que "crea" un objeto, todos los objetos de una misma clase tienen el mismo tipo de atributos
# que son sus caracteristicas, y acceso a los mismos metodos

# Puede ser util pensar en eso como una analogía de galletas, donde la clase es el molde con el que se cortan, y los atributos son
# el sabor, si tiene chispas de chocolate o no, etc. y los metodos son guardar la galleta, y comer la galleta.

class Galleta:

    def __init__(self, sabor, tiene_chispas, tamaño = "mediana"):

        self.sabor = sabor
        self.tiene_chispas = tiene_chispas
        self.tamaño = tamaño  # Todas las galletas son medianas a no ser que digamos lo contrario

    def __str__(self):
        # Es lo que ocurre si decimos "print(galleta)
        return ("Soy una falleta de sabor " + self.sabor + " y tamaño " + self.tamaño)

    def comer(self):
        print("me comieron :(")

    def higiene(self, otra_galleta):
        if otra_galleta.tiene_chispas == True and self.tiene_chispas == False:
            return "no me guardes con la otra galleta que me mancho de chocolate :(" #Por que return en vez de print?
        else:                                                                        #Por que esto puede usarse como funcion!
            return "si me guardas con la otra galleta, YO no me voy a manchar"
```

```python
galleta_1 = Galleta("chocolate", True)
galleta_2 = Galleta("cheesecake", False, "grande")

print(galleta_1)
print(galleta_2)

higiene_12 = galleta_1.higiene(galleta_2)
higiene_21 = galleta_2.higiene(galleta_1)

print(higiene_12)
print(higiene_21)
print(galleta_1.sabor)
```

## Tuplas

```python
#Son listas inmutables, muchos de los metodos que te enseñe con listas no aplican a tuplas, tienen su utilidad en bases de datos.

#Así se crean y leen

a = ('Chile', 2, 4.15, 'Agosto') #El ejemplo del apunte
print(a)
print(a[2])
```

```python
# Se le pueden alimentar tuplas a una función para que obtenga multiples valores, y puede entregar una tupla para entregar
# multiples valores (Una lista también puede, a este punto, podemos pensar en una tupla como en una lista con menos funciones)

def suma(edd):
    return edd[0] + edd[1]

tupla = (3, 5)
lista = [3, 5]

print(suma(tupla))
print(suma(lista))
```

```python
# También existen las named tuples, estas existen como tuplas a las que se les pueden asignar nombres, y nombres a sus atributos
# .... Como una clase, de verdad que las tuplas no sirven de mucho a esta altura.

#usaré el ejemplo del apunte

from collections import namedtuple

# Asignamos un nombre a la tupla (Register_type), y los nombres de los atributos que tendrá
ChileanRegister = namedtuple('ChileanRegister_type', ['RUT', 'name', 'age'])
# Vemos que podemos asignar los atributos como string, antes los asigno como una lista (se podrá como una tupla??)
MyRegister = namedtuple('MyRegister_type', 'unique_id, nickname, age')

# Instanciación e inicialización de la tupla
c1 = ChileanRegister('13427974-5', 'Christian', 20)
c2 = MyRegister(3, 'lucasvsj', 5)

print(c1) 
print(type(c1)) #literal una clase.
print(c2)
print(type(c2))
```
